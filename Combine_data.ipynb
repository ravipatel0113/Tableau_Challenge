```python
import pandas as pd
```

```python
data_2020_06 = "Data/JC-202006-citibike-tripdata.csv"
data_2020_07 = "Data/JC-202007-citibike-tripdata.csv"
data_2020_08 = "Data/JC-202008-citibike-tripdata.csv"
data_2020_09 = "Data/JC-202009-citibike-tripdata.csv"
data_2020_10 = "Data/JC-202010-citibike-tripdata.csv"
data_2020_11 = "Data/JC-202011-citibike-tripdata.csv"
data_2020_12 = "Data/JC-202012-citibike-tripdata.csv"
# data_2021_01 = "Data/JC-202101-citibike-tripdata.csv"
# data_2021_02 = "Data/JC-202102-citibike-tripdata.csv"
# data_2021_03 = "Data/JC-202103-citibike-tripdata.csv"
# data_2021_04 = "Data/JC-202104-citibike-tripdata.csv"
# data_2021_05 = "Data/JC-202105-citibike-tripdata.csv"
```

```python
all_data = [data_2020_06, data_2020_07, data_2020_08, data_2020_09, data_2020_10, data_2020_11, data_2020_12]
```

```python
data_list = []

for all_d in all_data:
    data_list.append(pd.read_csv(all_d));
```

```python
bike_data = pd.concat(data_list, sort=False)
bike_data
```
